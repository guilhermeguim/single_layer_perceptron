```python
import numpy as np
```

```python
#Simulando operador AND
# entradas = np.array([[0,0],[0,1],[1,0],[1,1]])
# saidas = np.array([0,0,0,1])

#Simulando operador OR
entradas = np.array([[0,0],[0,1],[1,0],[1,1]])
saidas = np.array([0,1,1,1])

#Simulando operador XOR !!!!!NÃO FUNCIONA, NÃO É LINEARMENTE SEPARAVEL!!!!!
# entradas = np.array([[0,0],[0,1],[1,0],[1,1]])
# saidas = np.array([0,1,1,0])

pesos = np.array([0.0,0.0])
taxaAprendizagem = 0.1
```

```python
def stepFunction(soma):
    if (soma>=1):
        return 1
    return 0
```

```python
def calculaSaida(registro): #(entrada,pesos)
   soma = 0
   for i in range(len(registro)):
      soma = soma + (registro[i]*pesos[i])
   #s = registro.dot(pesos) --> Método com melhor processamento
   return stepFunction(soma)
```

```python
def treinar():
    erroTotal = 1
    while (erroTotal !=0):
        erroTotal=0
        for i in range(len(saidas)):
            saidaCalculada = calculaSaida(np.asarray(entradas[i]))
            erro = abs(saidas[i] - saidaCalculada)
            erroTotal = erroTotal+erro
            for j in range(len(pesos)):
                pesos[j] = pesos[j] + (taxaAprendizagem * entradas[i][j] * erro)
                print("Peso atualizado:" + str(pesos[j]))
        print("Total de erros:"+str(erroTotal))
```

```python
treinar()
```
